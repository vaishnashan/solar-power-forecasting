```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\User\Desktop\solar_code\solarenergy.csv")

# Convert datetime
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')

# Set as index for time-series operations
df.set_index('Datetime', inplace=True)

df.head()
```

```python
# Check missing values
df.isna().sum()

# Percentage of missing values
(df.isna().sum() / len(df)) * 100
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['solar_mw'], color='orange')
plt.title("Solar Power Output Over Time")
plt.xlabel("Time")
plt.ylabel("Solar Power (MW)")
plt.show()
```

```python
# Drop non-numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Covariance
numeric_df.cov()
```

```python
numeric_df.hist(figsize=(15,10), bins=30, color='skyblue')
plt.suptitle("Feature Distributions", size=16)
plt.show()
```

```python
df['solar_mw'].resample('D').mean().plot(figsize=(15,5), title="Daily Average Solar Output")
plt.show()

df['solar_mw'].resample('M').mean().plot(figsize=(15,5), title="Monthly Average Solar Output")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['solar_mw'])
plt.title("Outliers in Solar Power")
plt.show()
```

```python
df['wind-direction'] = pd.to_numeric(df['wind-direction'], errors='coerce')

plt.figure(figsize=(10,6))
plt.scatter(df['wind-direction'], df['solar_mw'], alpha=0.5, color='teal')
plt.title("Wind Direction vs Solar Power")
plt.xlabel("Wind Direction")
plt.ylabel("Solar Power (MW)")
plt.show()
```

```python
rolling_corr = df['solar_mw'].rolling(500).corr(df['temperature'])

plt.figure(figsize=(12,5))
rolling_corr.plot()
plt.title("Rolling Correlation: Solar vs Temperature")
plt.ylabel("Correlation")
plt.show()
```
