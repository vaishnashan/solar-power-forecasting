```python
import pandas as pd

# Load cleaned files
gen = pd.read_csv(r"C:\Users\User\Desktop\solar_code\Task\plant2\Plant2_Generation_Cleaned.csv")
weather = pd.read_csv(r"C:\Users\User\Desktop\solar_code\Task\plant2\Plant2_Weather_Cleaned.csv")

# Ensure datetime consistency
gen['DATE_TIME'] = pd.to_datetime(gen['DATE_TIME'])
weather['DATE_TIME'] = pd.to_datetime(weather['DATE_TIME'])

# Merge on DATE_TIME
merged = pd.merge(gen, weather, on='DATE_TIME', how='inner')

print("Merged shape:", merged.shape)
print(merged.head())

merged.head(50)
```

```python
# Save merged dataset
merged.to_csv(
    r"C:\Users\User\Desktop\solar_code\Task\plant2\Plant2_Merged_Cleaned.csv",
    index=False
)

print("✅ Merged dataset saved successfully!")

import pandas as pd
```
