```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Plant 1 data
weather= pd.read_csv(r"C:\Users\User\Desktop\solar_code\Task\plant1\Plant_1_Weather_Sensor_Data.csv")

```

```python
weather
```

```python
# Drop unnecessary columns
weather_clean = weather.drop(columns=['PLANT_ID', 'SOURCE_KEY'])
```

```python
weather_clean
```

```python
# Ensure datetime
weather_clean['DATE_TIME'] = pd.to_datetime(weather_clean['DATE_TIME'])
```

```python
# Create full expected timeline (same as generation)
full_time_index= pd.date_range(
    start = weather_clean['DATE_TIME'].min(),
    end   = weather_clean['DATE_TIME'].max(),
    freq='15T'
)
 
len(full_time_index)
```

```python
# Reindex to full 15-min timeline
weather_clean = (
    weather_clean.set_index('DATE_TIME')
    .reindex(full_time_index)
    .reset_index()
    .rename(columns={'index':'DATE_TIME'})
)

print("Shape after reindexing:", weather_clean.shape)  # should be 3264
```

```python
# Interpolate missing values
weather_clean[['AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION']] = \
    weather_clean[['AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION']].interpolate(method='linear')
```

```python

# Final check
print(weather_clean.isnull().sum())
print(weather_clean.head(10))
print(weather_clean.tail(10))
```
