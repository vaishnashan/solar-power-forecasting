```python
from datasets import load_dataset
import pandas as pd
import numpy as np

ds = load_dataset("Creatorin/solar_selected")
```

```python
ds
```

```python

# use train split
train_df = ds["train"].to_pandas()
```

```python
train_df.shape
```

```python
train_df
```

```python

# drop index-like column if present
if "__index_level_0__" in train_df.columns:
    train_df = train_df.drop(columns=["__index_level_0__"])

# target and features
y = train_df["Leistung"]
X = train_df.drop(columns=["Leistung"])
```

```python
y
```

```python
X
```

```python



feature_names = X.columns.tolist()
print("Num features:", len(feature_names))
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
)

xgb.fit(X, y)

importances = xgb.feature_importances_
```

```python
k = 30  # choose how many you want to keep for CNN
idx = np.argsort(importances)[::-1][:k]

top_features = [feature_names[i] for i in idx]
top_importances = importances[idx]

print("Top features:")
for name, imp in zip(top_features, top_importances):
    print(f"{name:50s} {imp:.4f}")
```

```python
X_train_top = X[top_features].values  # shape (n_samples, k)
```

```python
X_train_top.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 10))

# reverse so most important is on top in horizontal bar plot
y_pos = np.arange(len(top_features))
plt.barh(y_pos, top_importances[::-1])
plt.yticks(y_pos, top_features[::-1])
plt.xlabel("XGBoost Feature Importance")
plt.title(f"Top {k} Features for Solar Power Forecasting")
plt.tight_layout()
plt.show()
```
