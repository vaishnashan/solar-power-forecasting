```python
import pandas as pd

# Load cleaned files
gen = pd.read_csv(r"C:\Users\User\Desktop\solar_code\Task\plant1\Plant1_Generation_Cleaned.csv")
weather = pd.read_csv(r"C:\Users\User\Desktop\solar_code\Task\plant1\Plant1_Weather_Cleaned.csv")
```

```python
# Ensure datetime consistency
gen['DATE_TIME'] = pd.to_datetime(gen['DATE_TIME'])
weather['DATE_TIME'] = pd.to_datetime(weather['DATE_TIME'])
```

```python
# Merge on DATE_TIME
merged = pd.merge(gen, weather, on='DATE_TIME', how='inner')
```

```python
print("Merged shape:", merged.shape)
print(merged.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix
corr_matrix = merged.corr(numeric_only=True)

# Print raw correlation values
print(corr_matrix)

# Heatmap visualization
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix - Plant 1 Merged Data", fontsize=14)
plt.show()
```
